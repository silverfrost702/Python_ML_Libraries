```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

# Plot Appearance

### Line Color 

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, color = 'black' )
```

 We can also use RGB hex codes for more custom colors.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, color = '#FF8C00' )    
```

### Line Width and Line Style

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, y, color = 'purple', linewidth = 3 )       # linewidth = 1 is default line width  
```

Alpha Arguement: It helps us to control the transparacy of the line.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                              # short hand for linewidth is lw
axes.plot(x, y, color = 'purple', lw = 3, alpha = 0.5 )     # helps us in getting the required shade 
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                              
#axes.plot(x, y, color = 'purple', lw = 3, linestyle = '-.' )
#axes.plot(x, y, color = 'purple', lw = 3, linestyle = ':' )
#axes.plot(x, y, color = 'purple', lw = 3, linestyle = 'steps' )
axes.plot(x, y, color = 'purple', lw = 3, ls = '--' )              # short hand for linestyle is ls
```

### Markers

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                                        # used when we have less data points
#axes.plot(x, y, color = 'purple', lw = 3, ls = '-', marker = 'o')     # all the 11 points present in x are marked
axes.plot(x, y, color = 'red', lw = 2, ls = '-', marker = '*')  
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                                        
axes.plot(x, y, color = 'purple', lw = 2, ls = '-', marker = '1', markersize = 20)  # used for better visibility of markers
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                                        
axes.plot(x, y, color = 'purple', lw = 5, ls = '-', marker = 'o', markersize = 20,
         markerfacecolor = 'yellow', markeredgewidth = 5, markeredgecolor = 'red')
```

## Control Over Axis Appearance

### Plot Range

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])                                        
axes.plot(x, y, color = 'purple', lw = 2, ls = '--')

axes.set_xlim([0,1])    # we give both upper and lower bound
axes.set_ylim([0,2])
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
x = np.arange(0,100)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y,color = '#A80617')     
ax2.plot(x,y, color ='#A80617')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_xlim([0,100])
ax.set_ylim([0,10000])

ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) 
axes[0].plot(x,y,lw = 3,ls = '--')
axes[1].plot(x,z,color = 'red',lw = 3,ls = '-')

axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,2.1))

axes[0].plot(x,y,color = '#2222D1',lw = 5,ls = '-')
axes[1].plot(x,z,color = 'red',lw = 3,ls = '--')

axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])
```
