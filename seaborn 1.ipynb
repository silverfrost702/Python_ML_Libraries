```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')       # there are some in-built datasets that can be loaded directly
```

```python
tips.head()
```

## 1) distplot()

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde = False)      # y axis shows the count and x axis shows the bars as bins.
```

histogram is essentially the distribution where the total bill lies.

```python
sns.kdeplot(tips['total_bill'])          # for kde plots
```

```python
sns.distplot(tips['total_bill'], kde = False, bins = 40 ) #we can change bin size according to use but it is good to choose a balanced bin sixe
```

## 2) jointplot()

```python
sns.jointplot(x = 'total_bill',y ='tip' ,data = tips)
```

From above we can see two distribution plots where tips is along the y-axis and total_bill along the x-axis. In the centre we have the scatter plot which basically shows the trend that if we go higher in total_bill then we will go higher in tip as well. 



Now we have a special parameter in jointplot() known as kind (that helps in comparing the distributions) which actually affects what actually happens in the joint plot. By default it is scatter we can change it to, for example hex.

```python
sns.jointplot(x = 'total_bill',y ='tip' ,data = tips, kind = 'hex')     #hex gives the hexagonal distribution. 
                                                                        #More no. of points more darker it gets. 
```

```python
sns.jointplot(x = 'total_bill',y ='tip' ,data = tips, kind = 'reg')   # regression basically shows linear equivalent of the 
                                                                      # scattered plot
```

```python
sns.jointplot(x = 'total_bill',y ='tip' ,data = tips, kind = 'kde')     # allows 2-D kde that shows where the density is more 
                                                                        # i.e. no. points are more (that reagon is darkest)
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'kde',color = 'purple')   #color as argument
```

## 3) pairplot()

NOTE: When a plot is across the same columns on both the axes, we see a histogram instead of the scattered plot.

```python
sns.pairplot(tips,hue='sex',palette = 'gnuplot2')   # changing the colors
```

## 5) Categorical Plots

The focus is on ploting catagorical data (from catagorical column(s)) with reference to either a numercial column or another categorical column.

### i) barplot()

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)   # allows to aggregate categorial data based on some function,
                                                        # by default mean();it is like visualization of group by
```

By default it shows mean or average of total_bill by catagorical column (eg. sex)

#### Note
If we want to affect the aggregate or average we can do that through estimator argument.

```python
# estimator :Statistical function to estimate within each categorical bin: by default it is mean()
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)   # it shows standard deviation of total_bill per category
```

For standard deviation function we first import numpy.

### ii) countplot()

```python
sns.countplot(x='sex',data=tips)  # explicit counting of no. of occurances. Here the y axis counts the occurances
```

### iii) boxplot()
It shows the distribution of quantitative data in a way that compares variables

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')  # it can split further by another categorical column 
```

### iv) violinplot()

```python
sns.violinplot(x='day',y='total_bill',data = tips)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips, hue = 'sex')
```

By split args we can directly compare distributions based on another category. Information on each half just doubles.

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips, hue = 'sex', split = True)
```

### v) stripplot()

```python
sns.stripplot(x = 'day',y = 'total_bill', data = tips, jitter = True)
```

```python
sns.stripplot(x = 'day',y = 'total_bill', data = tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)#use dodge in place of split to avoid warning
```

### vi) swarmplot()

```python
sns.swarmplot(x = 'day',y = 'total_bill', data = tips)   # combination of violin and strip plot
```

```python
sns.violinplot(x = 'day',y = 'total_bill', data = tips)
sns.swarmplot(x = 'day',y = 'total_bill', data = tips, color = 'pink')
```

### vii) factorplot()

```python
sns.factorplot(x = 'day',y = 'total_bill', data = tips, kind = 'bar')  # kind gives the type of plot
```

```python
sns.factorplot(x = 'day',y = 'total_bill', data = tips, kind = 'violin') 
```
