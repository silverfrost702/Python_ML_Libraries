## Functional Method

```python
plt.plot(x,y)
# plt.show() 
```

plt.show() can be used to display if we dont use jupyter otherwise not required

```python
plt.plot(x,y,'r-')   # color and linestyle 
```

#### If we want to label x,y and give title to our plot we can do this by xlabel(), ylabel(), title().

```python
plt.plot(x,y)             # plot made
plt.xlabel('X Label')     # label for x given as arguement
plt.ylabel('Y Label')     # label given as arguement
plt.title('My Plot')      # title is also given as arguement
#plt.show()               # used to display the plot(not compulsory in jupyter)
```

### Multiplots in same canvas

```python
plt.subplot(1,2,1)        # plot no.1
plt.plot(x,y,'r')
                          
plt.subplot(1,2,2)        #plot no.2
plt.plot(y,x,'b')         #subplot(rows,columns,plot no.)
```

Different colors and line styles

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')

plt.subplot(1,3,2)
plt.plot(y,x,'g*--')

plt.subplot(1,3,3)
plt.plot(x,y,'bo-')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r')

plt.subplot(2,1,2)
plt.plot(y,x,'b')
```

## Object Oriented

```python
fig = plt.figure()                        # creating figure object

axes = fig.add_axes([0.1,0.1,0.8,0.8])    # fig.add_axes(left,bottom,width,height)
axes.plot(x,y)                            # ploting on the axes
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])    
axes.plot(x,y) 

axes.set_xlabel('X Label')      
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

### Two Set of figures in one canvas

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes_2 = fig.add_axes([0.3,0.5,0.4,0.3])   # 30% from the left, 40% from the bottom and so on 
```

Therefore, we get insert set of axes

```python
fig = plt.figure()                           # we create an empty canvas

axes_1 = fig.add_axes([0.1,0.1,0.8,0.8])     # we add a set of axes to that empty canvas
axes_2 = fig.add_axes([0.2,0.5,0.4,0.3])   

axes_1.plot(x,y,'r')                         # we insert plots int the axes
axes_1.set_title('Plot 1')

axes_2.plot(y,x)
axes_2.set_title('Plot 2')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) 
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()                                # to overcome overlapping  of multiple hori. and verti. marking 
```

```python
axes    # array/list of matplotlib axes therefore iterable
```

### Iteration through axes

```python
fig,axes = plt.subplots(nrows = 1, ncols =2)
for current_ax in axes:
    current_ax.plot(x,y)
```

#### Axes is just a list of matplotlib axes(plural) which we create manually while writting axes = fig.add_axes([...]).

#### Difference between fig = plt.figure() and fig = plt.subplots() is that in plt.subplots() we automatically do add_axes([...]) based on no. of rows and no. of columns. 

### Indexing through axes

```python
fig,axes = plt.subplots(nrows =1,ncols =2)
axes[0].plot(x,y)
axes[1].plot(y,x,'g')
```

```python
fig,axes = plt.subplots(nrows =1,ncols =2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('X axis')
axes[0].set_ylabel('Y axis')

axes[1].plot(y,x,'g')
axes[1].set_title('Second Plot')
axes[1].set_xlabel('X axis')
axes[1].set_ylabel('Y axis')

plt.tight_layout()    # for clean representation
```

## Figure Size, Aspect Ratio and DPI

```python
fig = plt.figure(figsize = (15,5))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,2.2))
axes[0].plot(x,y,'g')
axes[0].set_title('FIRST')

axes[1].plot(y,x)
axes[1].set_title('SECOND')

plt.tight_layout()
```

## Legend()

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.plot(y,x)

axes.legend()
```

Therefore we need to mention the labels of the axes while plotting.

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,label = 'curve 1')
axes.plot(y,x,label = 'curve 2')

axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label = 'square')
axes.plot(x,x**3,label = 'cube')

axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,label = 'square')
axes.plot(x,x**3,label = 'cube')

#axes.legend(loc = 0)          # loc from 0-10 are 10 specific locations for legend
#axes.legend(loc = 2)
#axes.legend(loc = 4)
axes.legend(loc = (0.1,0.7))   # custom made location can be made by giving a tuple as an arg. with x and y cooridinates
```
